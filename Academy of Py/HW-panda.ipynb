```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of students

* Calculate the average math score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Create a dataframe to hold the above results

```python
# District Summary dataframe
D_summary = pd.DataFrame()
D_summary["Total Schools"] =[school_data.school_name.nunique()]
D_summary["Total Students"] = [student_data["Student ID"].nunique()]
D_summary["Total Budget"]=[school_data["budget"].sum()]
D_summary["Average Math Score"]=[student_data["math_score"].mean()]
D_summary["Average Reading Score"]=[student_data["reading_score"].mean()]
D_summary["Overall Average Score"]=(D_summary["Average Reading Score"]+D_summary["Average Math Score"])/2
D_summary["Percentage Passing Math"]=([100*(student_data["math_score"] >= 70).mean()])
D_summary["Percentage Passing Reading"]=([100*(student_data["reading_score"] >= 70).mean()])
D_summary["Overall Passing Rate"]=(D_summary["Percentage Passing Math"] +D_summary["Percentage Passing Reading"])/2
D_summary
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
# School Summary dataframe
S_summary = pd.DataFrame()
S_summary["Type"] = school_data_complete.groupby("school_name")['type'].unique()
S_summary["Total Students"] = school_data_complete.groupby("school_name")['Student ID'].nunique()
S_summary["Total Budget"] = school_data_complete.groupby("school_name")['budget'].mean()
S_summary["Average Math Score"] = school_data_complete.groupby("school_name")['math_score'].mean()
S_summary["Average Reading Score"] = school_data_complete.groupby("school_name")['reading_score'].mean()
S_summary["Overall Average Score"]=(S_summary["Average Reading Score"]+S_summary["Average Math Score"])/2
S_summary["Percentage Passing Math"]= 100 * school_data_complete[school_data_complete['math_score'] >= 70].groupby("school_name")['Student ID'].count()/S_summary["Total Students"]
S_summary["Percentage Passing Reading"]=100 * school_data_complete[school_data_complete['reading_score'] >= 70].groupby("school_name")['Student ID'].count()/S_summary["Total Students"]
S_summary["Overall Passing Rate"] = (S_summary["Percentage Passing Reading"] + S_summary["Percentage Passing Math"])/2



S_summary.sort_values(by=['Overall Passing Rate'], ascending=False).head(5)
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
#creates grade level average math scores for each school 
math_scores = pd.DataFrame()

math_scores["9th"] = student_data.loc[student_data['grade'] == '9th'].groupby('school_name')["math_score"].mean()
math_scores["10th"] = student_data.loc[student_data['grade'] == '10th'].groupby('school_name')["math_score"].mean()
math_scores["11th"] = student_data.loc[student_data['grade'] == '11th'].groupby('school_name')["math_score"].mean()
math_scores["12th"] = student_data.loc[student_data['grade'] == '12th'].groupby('school_name')["math_score"].mean()

math_scores
```

## Reading Score by Grade 

* Perform the same operations as above for reading scores

```python
#creates grade level average math scores for each school 
reading_scores = pd.DataFrame()

reading_scores["9th"] = student_data.loc[student_data['grade'] == '9th'].groupby('school_name')["reading_score"].mean()
reading_scores["10th"] = student_data.loc[student_data['grade'] == '10th'].groupby('school_name')["reading_score"].mean()
reading_scores["11th"] = student_data.loc[student_data['grade'] == '11th'].groupby('school_name')["reading_score"].mean()
reading_scores["12th"] = student_data.loc[student_data['grade'] == '12th'].groupby('school_name')["reading_score"].mean()

reading_scores
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
school_data_complete['spending_bins'] = pd.cut(school_data_complete['budget']/school_data_complete['size'], spending_bins, labels = group_names)


sp_bin_summary = pd.DataFrame()
sp_bin_summary["Average Math Score"] = school_data_complete.groupby("spending_bins")['math_score'].mean()
sp_bin_summary["Average Reading Score"] = school_data_complete.groupby("spending_bins")['reading_score'].mean()
sp_bin_summary["Percentage Passing Math"]= 100 * school_data_complete[school_data_complete['math_score'] >= 70].groupby("spending_bins")['Student ID'].count()/school_data_complete.groupby("spending_bins")['Student ID'].nunique()
sp_bin_summary["Percentage Passing Reading"]=100 * school_data_complete[school_data_complete['reading_score'] >= 70].groupby("spending_bins")['Student ID'].count()/school_data_complete.groupby("spending_bins")['Student ID'].nunique()
sp_bin_summary["Overall Passing Rate"] = (sp_bin_summary["Percentage Passing Reading"] + sp_bin_summary["Percentage Passing Math"])/2

sp_bin_summary
```

## Scores by School Size

```python
# Sample bins. Feel free to create your own bins.
size_bins = [0, 1000, 2000, 5000]
group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

school_data_complete['size_bins'] = pd.cut(school_data_complete['size'], size_bins, labels = group_names)


size_bin_summary = pd.DataFrame()
size_bin_summary["Average Math Score"] = school_data_complete.groupby("size_bins")['math_score'].mean()
size_bin_summary["Average Reading Score"] = school_data_complete.groupby("size_bins")['reading_score'].mean()
size_bin_summary["Percentage Passing Math"]= 100 * school_data_complete[school_data_complete['math_score'] >= 70].groupby("size_bins")['Student ID'].nunique()/school_data_complete.groupby("size_bins")['Student ID'].nunique()
size_bin_summary["Percentage Passing Reading"]=100 * school_data_complete[school_data_complete['reading_score'] >= 70].groupby("size_bins")['Student ID'].nunique()/school_data_complete.groupby("size_bins")['Student ID'].nunique()
size_bin_summary["Overall Passing Rate"] = (size_bin_summary["Percentage Passing Reading"] + size_bin_summary["Percentage Passing Math"])/2

size_bin_summary
```
